```python
import numpy as np

# evenly sampled time at 200ms intervals
t = np.arange(0., 5., 0.2)

plt.figure()

plt.subplot(3 ,1 , 1)
plt.xlabel("t value")
plt.ylabel("Output Value")
plt.plot(t, t, 'r--')
plt.title("Linear")

plt.subplot(3 , 1 , 2)
plt.plot( t,t**2, 'bs')
plt.xlabel("t value")
plt.ylabel("Output Value")
plt.title("square")

plt.subplot(3 , 1 , 3)
plt.plot( t, t**3, 'g^')
plt.xlabel("t value")
plt.ylabel("Output Value")
plt.title("cube")

plt.suptitle("Main titile")
plt.tight_layout()# for maintain good spacing

plt.show()
```

```python
# axes - x axis , y axis , labels ,tickers

x = np.linspace(1 , 10 , 100)
print(x)
fig , axes = plt.subplots(2 , 3)

axes[0,0].plot(x , np.sin(x))
axes[0,0].set_title("Sin function")
axes[0,0].set_xlabel("x")
axes[0,0].set_ylabel("y")

axes[0,1].plot(x , np.exp(x))
axes[0,1].set_title("Expo function")
axes[0,1].set_xlabel("x")
axes[0,1].set_ylabel("y")

axes[0,2].plot(x , np.tan(x))
axes[1,0].plot(x , np.sqrt(x))
axes[1,1].plot(x , np.log(x))
axes[1,2].plot(x , np.pow(x , 2))


plt.tight_layout()
```

```python
courses = ["DSA" , "Python" , "ML" , "AI"]
students = [2000 , 1800 , 2500 , 1500]

plt.bar(courses , students, width=0.5 , align='edge' , edgecolor='black' , linewidth=4 , color='yellow' , hatch="*")
plt.show()
```

```python
courses = ["DSA" , "Python" , "ML" , "AI"]
students = [2000 , 1800 , 2500 , 1500]
hatches = ['/' , '*' , 'o' , '\\']
bar_colors = ["red" , "yellow" , "m" , "pink"]

bars = plt.bar(courses , students, width=0.5 , align='edge' , edgecolor='black' , linewidth=4 , color='yellow' )

for bar , hatch , color in zip(bars , hatches , bar_colors):
    bar.set_color(color)
    bar.set_hatch(hatch)
    bar.set_edgecolor("k")

    
plt.show()
```

```python
### pi chart
courses = ["DSA" , "Python" , "ML" , "AI" , "HLD"]
students = [2000 , 1800 , 2500 , 1500 , 2000]
pie_colors = ["red" , "yellow" , "m" , "pink" , "blue"]

explode = [ 0.2 , 0 , 0 , 0 , 0]
plt.pie(students , labels=courses , shadow=True  , explode=explode , autopct="%.2f%%" , colors=pie_colors)
plt.legend(loc="upper left")
plt.show()
```

```python
# 3d scatter plotab

ax = plt.axes(projection='3d')
x = np.random.rand(100)
y = np.random.rand(100)
z = np.random.rand(100)

ax.scatter(x, y , z)
ax.set_xlabel('x')
ax.set_ylabel('y')
ax.set_zlabel('z')

plt.show()
```

```python
# 3d plot
ax = plt.axes(projection='3d')

x = np.linspace(0 , 20 , 100)
y = np.sin(x)
z = np.cos(x)

ax.plot(x ,y , z)
plt.show()
```

```python
# 3d bar 
ax = plt.axes(projection='3d')

x = np.random.randint(1 , 10  , 5)
y = np.random.randint(1 , 10  , 5)
z = np.zeros(5)

dx = np.ones(5)
dy = np.ones(5)
dz = np.random.randint(1 , 10 ,5)

ax.bar3d(x , y , z , dx , dy , dz)

plt.show()
```
